```python
def plot_mean_and_CI(arrays,color_mean,color_shading,patch=True):
    mean=[array[0] for array in arrays]
    ub=[array[1][0] for array in arrays]
    lb=[array[1][1] for array in arrays]
    # plot the shaded range of the confidence intervals
    if patch==True:
        plt.fill_between(range(len(mean)), ub,lb,color=color_shading, alpha=.5)
    plt.plot(mean,color_mean)
    
```
